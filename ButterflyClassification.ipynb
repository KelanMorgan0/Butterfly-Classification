# Butterfly Classification
the purpose of this jupyter notebook is to classify butterfly's by Images, using k-nearest neighbour

# Imports

```python
import os
import cv2
import numpy as np
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report
from skimage.feature import hog
from sklearn.preprocessing import LabelEncoder
```

# Load the Data

```python
data = pd.read_csv("Data/butterflies and moths.csv")
```

# Function to check the image

```python
def load_image_from_path(file):
    if os.path.exists(file):
        #open the image
        image = cv2.imread(file)

        #Shrink the image to a size that is easier to handle
        image = cv2.resize(image, (64, 64))

        #make the pixels have a value from 0 to 1
        image = image / 255.0
        return image
    else:
        return None
```

# Load all images and lables into a list

```python
images = []
labels = []

for index, row in data.iterrows():
    image_path = row['filepaths']
    image = load_image_from_path("Data/" + image_path)
    if image is not None:
        images.append(image)
        labels.append(row['labels'])

#make the lists numpy arrays
images = np.array(images)
labels = np.array(labels)
```

# Use Histogram of Oriented Gradients (Hog) to achieve better accuracy
gets edge and texture information and colour information in a split up image

```python
def getHogImages(image):
    fd, hog_image = hog(image, pixels_per_cell=(8, 8), cells_per_block=(2, 2), visualize=True, channel_axis=-1)
    return fd

images = np.array([getHogImages(image) for image in images])

# Encode labels for the model to use
label_encoder = LabelEncoder()
labels = label_encoder.fit_transform(labels)
```

# Split the data in to test and training

```python
train_mask = data['data set'] == 'train'
X_train = images[train_mask]
y_train = labels[train_mask]
X_test = images[~train_mask]
y_test = labels[~train_mask]
```

# Train the model
specify the number of neiborus to check

```python
model = KNeighborsClassifier(n_neighbors=1)
model.fit(X_train, y_train)
```

# run the model using the test data and predict accuracy

```python
y_pred = model.predict(X_test)

# Calculate accuracy
print(classification_report(y_test, y_pred))
```
